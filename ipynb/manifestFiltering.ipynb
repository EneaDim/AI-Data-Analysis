# FIND CASE_ID IN mRNA JSON THAT EXIST ALSO IN IMAGE JSON AND TAKE ITS FILE NAME

## HEALTHY

```
file_in_json_mrna = open("/content/drive/My Drive/Bioinfo/Healthy/mRNA/files.2020-09-18.json","r+")
flag_case_id_mrna=0                                                             # FLAG CASE ID 
case_id_Str="case_id"                                                           # CASE ID STRING
case_id_list=list()                                                             # CASE ID's LIST
for line in file_in_json_mrna:                                                  # READ LINE OF JSON FILE
  string = line.split()                                                         # SPLIT LINE
  for substring in string:                                                      # READ THE SUBSTRING
    if case_id_Str in substring:                                                # IF = "case_id"
      flag_case_id_mrna=1                                                       # FLAG FOR READ THE NEXT SUBSTRING THAT IS THE CASE ID
    else:                                                                       # IF != "case_id"
      if (flag_case_id_mrna==1):                                                # RIGHT STRING FOR CASE ID
        flag_case_id_mrna=0                                                     # RESET FLAG
        case_id_list.append(substring)                                          # APPEND CASE ID 
print(case_id_list)
file_in_json_mrna.close()                                                       # CLOSE FILE
# FILTER MRNA SAMPLES ON ITS JSON FILE 
k=0
file_in_json_image = open("/content/drive/My Drive/Bioinfo/Healthy/Image/files.2020-08-26.json","r+")
flag_case_match=0                                                               # FLAG CASE MATCH
flag_case_id_image=0                                                            # FLAG CASE ID IMAGE
flag_case_filename=0                                                            # FLAG CASE FILENAME
filename_list=list()                                                            # FILENAME LIST
finalFilename_list=list()                                                       # FINAL FILENAMELIST
filename_id_Str="file_name"                                                     # 'filename' STRING
for line in file_in_json_image:                                                 # READ LINE IN JSON IMAGE FILE
  string = line.split()                                                         # SPLIT LINE
  for substring in string:                                                      # SUBSTRING IN STRING
    if case_id_Str in substring:                                                # IF = "case_id"
      flag_case_id_image=1                                                      # FLAG FOR READ THE NEXT SUBSTRING THAT IS THE CASE ID
    else:                                                                       # IF != "case_id"
      if (flag_case_id_image==1):                                               # RIGHT STRING FOR CASE ID
        flag_case_id_image=0                                                    # RESET FLAG
        for _id in case_id_list:                                                # ITERATE OVER CASE ID's LIST
          if substring in _id:                                                  # SUBSTRING = CASE ID IN CASE ID's
            k+=1                                                                # K MATCH
            flag_case_match=1                                                   # FLAG CASE MATCH
      if (flag_case_match==1):                                                  # IF CASE MATCH
        if (flag_case_filename==1):                                             # IF FLAG FILENAME MATCH
          flag_case_filename=0                                                  # RESET FLENAME MATCH FLAG
          flag_case_match=0                                                     # RESET CASE MATCH FLAG
          flag_case_id_image=0                                                  # RESET CASE ID IMAGE FLAGE
          filename_list.append(substring)                                       # APPEND THE CORRECT FILE NAME SUBSTRING
        if filename_id_Str in substring:                                        # IF SUBSTRING = 'filename'
          flag_case_filename=1                                                  # FLAG FOR READ THE NEXT SUBSTRING THAT IS THE FILENAME
file_in_json_image.close()                                                      # CLOSE JSON FILE
for filename in filename_list:                                                  # ITERATE OVER ALL FILENAMES
  finalFilename=filename[1:65]                                                  # SELECT THE CORRECT STRING
  finalFilename_list.append(finalFilename)                                      # APPEND TO FINAL FILENAME LIST
print(finalFilename_list)                                   
# WRITE THE NEW MANIFEST FILE
file_in_manifest_image=open("/content/drive/My Drive/Bioinfo/Healthy/Image/gdc_manifest.2020-08-26.txt","r+")
file_out_manifest_image=open("/content/drive/My Drive/Bioinfo/Healthy/Image/gdc_manifest_image.txt","w+")
for line in file_in_manifest_image:                                             # ITERATE OVER MANIFET LINES
  if (line[0]=="i" and line[1]=="d"):                                           # IF FIRST LINE
    file_out_manifest_image.write(line)                                         # WRITE FIRST LINE
  print(line)
  for filename in finalFilename_list:                                           # ITERATE OVER FINAL FILENAME LIST
    if filename in line:                                                        # IF FILENAME IS IN THIS LINE          
      file_out_manifest_image.write(line)                                       # WRITE THE LINE
file_in_manifest_image.close()                                                  # CLOSE THE MANIFEST INPUT FILE
file_out_manifest_image.close()                                                 # CLOSE THE MANIFEST OUTPUT FILE
```
